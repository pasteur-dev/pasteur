```python
import os
```

```python
from zipfile import ZipFile

files = os.listdir("../raw/sdgym")
names = {fn.split("_v")[0].split(".zip")[0].lower(): fn for fn in files if fn.endswith(".zip")}

tables = []
for ds_name, ds_fn in names.items():
    with ZipFile(os.path.join("../raw/sdgym", ds_fn), "r") as zip:
        ds_tables = []
        for file in zip.filelist:
            fn = file.filename
            if not fn.endswith(".csv"):
                continue

            name = fn.split("/")[-1][:-4].lower()
            full_name = f"{ds_name}.{name}"
            ds_tables.append(full_name)
        
        if len(ds_tables) == 1:
            ds_tables = [f"{ds_name}.table"]

        tables.extend(ds_tables)

tables
```
