```python
base_loaded = {key: load() for i, (key, load) in enumerate(base.items()) if i < 7}
```

```python
dfs = {}

# Load old format base files
for key, tmp in base_loaded.items():
    print(f"Processing {key}...")
    # tmp = load()

    # Generate spec unit in new format
    spec_unit = tmp["spec_unit"].astype(str)
    get_spec_idx = (
        lambda i: spec_unit.str.slice(i, i + 1).replace("", pd.NA).astype("category")
    )
    spec_units = pd.DataFrame({f"spec_unit_{i + 1}": get_spec_idx(i) for i in range(5)})

    # Generate intermediary table
    # new base files now have normalized facility files, so drop denormalized
    # facility columns, as well as old `spec_unit`
    itrm = tmp.drop(
        columns=[
            "fac_acute_care_ind",
            "fac_long_term_ac_ind",
            "fac_other_ltc_ind",
            "fac_peds_ind",
            "fac_psych_ind",
            "fac_rehab_ind",
            "fac_snf_ind",
            "fac_teaching_ind",
            "spec_unit",
        ]
    ).join(spec_units)

    dfs[key] = itrm

dfs_old = dfs
```
